#**Numpy Indexing and Selection**

```python
arr = np.arange(0,11)
print(arr)
```

```python
arr[1]
```

```python
arr[:2]
```

```python
arr[:]
```

```python
arr[-3:-1]
```

```python
arr[0:5] = 100  # brodcasting
print(arr)
```

```python
arr = np.arange(0,11)
print(arr)
```

```python
slice_of_arr = arr[:6]  # this is a pointer to the actual array
print(slice_of_arr)
```

```python
slice_of_arr[:] = 99
print(slice_of_arr)
```

```python
print(arr) # as slice_of_array was a pointer
```

```python
arr_copy = arr.copy()
arr_copy[:] = 200
print(arr_copy)
```

```python
print(arr)
```

```python
arr = np.arange(1,11)
print(arr)
```

```python
arr > 4
```

```python
bool_arr = arr > 4
print(arr[bool_arr])
```

```python
print(arr[arr > 4])  # conditional selection
```

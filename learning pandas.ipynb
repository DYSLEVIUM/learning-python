# **Multi-Index and Index Heirarchy**

```python
outside = ['G1', 'G1', 'G1', 'G2', 'G2', 'G2']
inside = [1,2,3,1,2,3]
heir_index = list(zip(outside,inside))  # zip makes tuple pairs
heir_index = pd.MultiIndex.from_tuples(heir_index)
heir_index
```

```python
df = pd.DataFrame(randn(6,2),heir_index,['A','B'])
df
```

```python
print(df.index.names)
df.index.names = ['Groups','Num']
print(df.index.names)
```

```python
df
```

```python
df.loc['G1']
```

```python
df.xs('G1') # cross-section
```

```python
df.xs(1,level='Num') # all key values of a particular level
```

# **Missing Data**

```python
d = {'A':[1,2,np.nan],'B':[5,np.nan,np.nan],'C':[1,2,3]}
df = pd.DataFrame(d)
df
```

```python
df.dropna(axis=0,inplace=False, thresh=2) # removes any rows with one or more NAN, use inplace
df
```

```python
df.fillna(value='FILL VALUE') # filling missing data
```

```python
df['A'].fillna(value=df['A'].mean())  # filling missing data with the mean of the rows' value
```

# **Group By**

```python
# groupby allows us to group together rows based off of a column and perfoem an aggregate function on them
```

```python
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
        'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
        'Sales':[200,120,340,124,243,350]}
df = pd.DataFrame(data)
df
```

```python
byComp = df.groupby('Company')
```

```python
byComp.mean()
```

```python
byComp.sum()
```

```python
byComp.std()
```

```python
byComp.sum().loc['FB']
```

```python
df.groupby('Company').sum().loc['FB'] # one-liner
```

```python
df.groupby('Company').count()
```

```python
df.groupby('Company').max()
```

```python
df.groupby('Company').describe().transpose()['FB']
```

# **Concatenating, Merging and Joining**

```python
# concatenating
```

```python
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3'],
                    'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']},
                   index=[0,1,2,3])
```

```python
df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
                    'B':['B4','B5','B6','B7'],
                    'C':['C4','C5','C6','C7'],
                    'D':['D4','D5','D6','D7']},
                   index=[4,5,6,7])
```

```python
df3 = pd.DataFrame({'A':['A8','A9','A10','A11'],
                    'B':['B8','B9','B10','B11'],
                    'C':['C8','C9','C10','C11'],
                    'D':['D8','D9','D10','D11']},
                   index=[8,9,10,11])
```

```python
pd.concat([df1,df2,df3])  # dimesions should match along the axis of concatenation
```

```python
pd.concat([df1,df2,df3], axis=1)
```

```python
left = pd.DataFrame({'key':['K0','K1','K2','K3'],
                     'A':['A0','A1','A2','A3'],
                     'B':['B0','B1','B2','B3']})

right = pd.DataFrame({'key':['K0','K1','K2','K3'],
                     'C':['C0','C1','C2','C3'],
                     'D':['D0','D1','D2','D3']})
```

```python
left
```

```python
right
```

```python
# merging
```

```python
pd.merge(left, right, how='inner', on='key')
```

```python
# joining
```

```python
left = pd.DataFrame({'A':['A0','A1','A2'],
                     'B':['B0','B1','B2']},
                    index=['K0','K1','K2'])

right = pd.DataFrame({'C':['C0','C1','C2'],
                     'D':['D0','D1','D2']},
                     index=['K0','K1','K2'])
```

```python
left.join(right) #  same as merge but merged on the basis of index
```

```python
left.join(right, how='outer')
```
